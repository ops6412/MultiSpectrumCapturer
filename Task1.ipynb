```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Read in any image
image_path = 'Comp.jpg'  # Replace with your image path
original_image = cv2.imread(image_path)

# Step 2: Convert the image to YUV
yuv_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2YUV)

# Step 3: Extract the Y channel (luminance)
y_channel = yuv_image[:, :, 0]  # Y channel

# Step 4: Convert the original image to grayscale (intensity)
gray_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)

# Step 5 & 6: Show the images using Matplotlib
plt.figure(figsize=(15, 7))

# Subplot 1: Original Image
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for display
plt.title('Original Image')
plt.axis('off')

# Subplot 2: Y Channel (Luminance)
plt.subplot(1, 3, 2)
plt.imshow(y_channel, cmap='gray')
plt.title('Y Channel (Luminance)')
plt.axis('off')

# Subplot 3: Grayscale Image (Intensity)
plt.subplot(1, 3, 3)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image (Intensity)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
